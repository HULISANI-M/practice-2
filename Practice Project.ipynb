```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('car_features_and_msrp.CSV')
```

```python
data.head()  # Display the first few rows of the DataFrame
```

```python
data.info()  # Display information about the DataFrame, like data types and missing values
```

```python
data.describe()  # Summary statistics of numeric columns
```

```python
from scipy.stats import ttest_ind

# Example: T-test for two independent samples
group1 = data[data['MSRP'] == 'Group 1']['Number of Doors']
group2 = data[data['MSRP'] == 'Group 2']['Number of Doors']

t_stat, p_value = ttest_ind(group1, group2)
print(f"T-statistic: {t_stat}, p-value: {p_value}")
```

```python
# Example: Histogram of a numeric column
plt.hist(data['Engine Cylinders'])
plt.title('Histogram ')
plt.xlabel('Number of Doors')
plt.ylabel('highway MPG')
plt.show()
```

```python
# Example: Correlation matrix
correlation_matrix = data.corr()
print("Correlation Matrix:\n", correlation_matrix)

# Visualize the correlation matrix
import seaborn as sns
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

```python
import seaborn as sns

# Example: Pairplot for multiple variables
sns.pairplot(data[['city mpg', 'highway MPG', 'MSRP']])
plt.show()
```
